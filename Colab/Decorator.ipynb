# Function as parameter. Modularization

```python
import numpy as np
def func_bankac1(num):
  print (num**2)
  return num**2

def func_bankac2(num):
  print (num**3)
  return num**3

def func_bankac3(num):
  print (num**4)
  return num**4

def func2(func, numbers):
  # Some processes
  val = func(numbers)

  # Some processes
  val = val+np.log(val)
  return val

print("Finally Value for BankAc1 " , func2(func_bankac1, numbers=4))
print("Finally Value for BankAc2" , func2(func_bankac2, numbers=4))
print("Finally Value for BankAc3" , func2(func_bankac3, numbers=4))


```

# Function as an argument

```python
def inc(x):
    return x+1

def dec(x):
    return x-1

def operate(func, x):
    r = func(x)
    return r

#inc function is passed a parameter/argument
print ( operate(inc, 3) )

#dec function is passed a parameter/argument
print ( operate(dec, 15))
```

# Decorator

```python
def make_pretty(func):
    def inner():
        print ("I am decorated")
        func()
    return inner

#Decorator Function
def ordinary():
    print ("I am ordinary")
    
print (type(make_pretty))
print (type(ordinary))
```

```python
#Generally old function is decorated and get reassigned like this. ordinary = make_pretty(ordinary)
ordinary = make_pretty(ordinary)
ordinary()
```

```python
#Not ordinary is still a function
print(ordinary)
```

### A better of doing above decoration work in python is as below.

```python
def make_pretty(func):
    def inner():
        print ("I am decorated")
        func()
    return inner

@make_pretty
def ordinary():
    print ("I am ordinary")
    
ordinary()
```

```python
#Ordinary is not a function but a decorator. Therefore type is NoneType
type(ordinary())
```

### Decorator Use- Example 2

```python
def smart_divide(func):
    def inner(a, b):
        print("I am going to divide", a, "and", b)
        if b == 0:
            print("Whoops! cannot divide")
            return

        return func(a, b)
    return inner


@smart_divide
def divide(a, b):
    print(a/b)
    
divide(5,2)
print ("*********")
divide(5,0)
```

### Decorator Example 3

```python
def star(func):
    def inner(*args, **kwargs):
        print("*" * 30)
        func(*args, **kwargs)
        print("*" * 30)
    return inner


def percent(func):
    def inner(*args, **kwargs):
        print("%" * 30)
        func(*args, **kwargs)
        print("%" * 30)
    return inner


@star
@percent
def printer(msg):
    print(msg)


printer("Hello")
```

```python
@percent
@star
def printer(msg):
    print(msg)


printer("Hello")
```

```python
def func1(x):
    print ("my name is func1")
    def func11(*args, **kwargs):
        print("func11 decorator" +"*"* 5)
        print ("1st Time")
        x(*args, **kwargs)
        print ("2nd Time")
        x(*args, **kwargs)
        print ("3rd Time")
        x(*args, **kwargs)
        print("func11 decorator"+"*"* 10)
    return func11

def func2(y):
    print ("my name is func2")
    def func21(*args, **kwargs):
        print("\t func21 decorator"+"%"* 15)
        y(*args, **kwargs)
        print("\t func21 decorator"+"%"* 25)
    return func21

def func3(z):
    print ("my name is func3")
    def func31(*args, **kwargs):
        print("\t\t func31 decorator"+"#"* 15)
        z(*args, **kwargs)
        print("\t\t func31 decorator"+"#"* 25)
    return func31

def func4(a):
    print ("my name is func4")
    def func41(*args, **kwargs):
        print("\t\t\t func41 decorator"+"#"* 15)
        a(*args, **kwargs)
        print("\t\t\t func41 decorator"+"#"* 25)
    return func41


import numpy as np

@func1
@func2
@func3
@func4
def decorate_me(m):
    print("\t\t\t\t "+m)  
    n=13
    a=np.sqrt(n)
    print (f"\t\t\t\t sqrt of {n} is {a}")
    print("\t\t\t\t Do this also")
    
## Note without call of any function these four lines are being executed from each decorator function.
## Note they are bring printed in reverse order
## Note all 4 decorators are independt functions and their input is function decorate_me alongwith parameters.
## It means x,y,z,a of func1, func2, func3, func4 is nothing but decorate_me
## Because all 4 are decorator function and in our case all decorating the same function i.e. decorate_me
```

```python
def deco_func1(func):
  def wrapper1():
    print("deco_func1 wrapper")
    func()
    
  return wrapper1 

def deco_func(func):
  def wrapper():
    print ("Line before call")
    func()
    print ("Line after call")
  return wrapper

@deco_func1
@deco_func
def say_hello():
  print ("Hello Hari")
  return

say_hello()
```
