```python
import pandas as pd

# Load channel on browser. Goto the bottom of the page (sometimes there are hundreds of videos which are not loaded on one page first time)
# copy the title of the last of video from the page
# right click on the page and click inspect
# search this last video text in the inspect window
# Move one step up in the hierarchy of the html where this text is found. For this you need to find closing tag of the previous tag.
# Select the tag. Right click and click edit.
# copy the html tag.
# create a file like lexfridman_channel.html in your current working folder.
# paste all the text and save the file.
```

```python
from google.colab import files

with open('/content/lexfridman_channel.html', 'r') as f:
  content=f.readlines()
```

```python
df=pd.DataFrame({"Content": content})

df.shape
```

```python
df=df.sort_values('Content')

df=df.drop_duplicates()

df.shape
```

```python
df['Content']=df['Content'].str.replace('"','')

# keep only those rows which has youtube video information.
df = df[df['Content'].str.find('views title=')>0]
df
```

```python
#get the title from the content. It will have title plus href info.
df['title'] = df['Content'].str.split('views title=').apply(lambda x: x[1])

#get the href from title
df['href']=df['title'].str.split('href').apply(lambda x: x[1])
df['href']=df['href'].str.split('><yt-formatted').apply(lambda x: x[0])

#update the url correctly
df['href'] = 'https://www.youtube.com/' + df['href'].str[3:]

#remove other information from title
df['title']=df['title'].str.split('href').apply(lambda x: x[0])

#keep only title and href in dataset
df = df.loc[:,['title','href']]

#during sorting process, order of row changed. Bring the same order rows which as in original file.
df = df.sort_index()

#Save the file
df.loc[:,['title','href']].to_csv('lexfridman_channel.csv', index=None)
```
