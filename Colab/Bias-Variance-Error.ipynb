```python
import numpy as np

np.random.seed(123)

n_samples = 100
Emp_Age = np.random.normal(35, 5, n_samples)
Emp_Exp = np.random.normal(10, 2, n_samples)
Expected_Salary = 30000 + 500 * Emp_Age + 1000 * Emp_Exp + np.random.normal(0, 1000, n_samples)

data = np.column_stack([Emp_Age, Emp_Exp, Expected_Salary])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data[:, :2], data[:, 2], test_size=0.2, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
n_runs = 50
n_train = 80

mse = np.zeros(n_runs)

for i in range(n_runs):
    train_idx = np.random.choice(n_samples, n_train, replace=False)
    X_train_i =data[:, :2][train_idx]
    y_train_i = data[:, 2][train_idx]
    lr_i = LinearRegression()
    lr_i.fit(X_train_i, y_train_i)
    y_pred_i = lr_i.predict(X_test)
    mse[i] = np.mean((y_test - y_pred_i) ** 2)

bias_error = (np.mean(y_test) - np.mean(lr.predict(X_test))) ** 2
variance_error = np.mean(mse) - bias_error
total_error = np.mean(mse)

print("Bias Error:", bias_error)
print("Variance Error:", variance_error)
print("Total Error:", total_error)
```
