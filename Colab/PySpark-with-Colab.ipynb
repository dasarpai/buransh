```python
df1 = spark.read.csv('/content/shampoo.csv', header=True)
```

```python
df1 = spark.table("shampoo.csv")
display(df1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dataset = df1.toPandas()
total_data = df1["Month"].count()
split = (total_data * 0.90).astype(np.int32)
train = dataset[0:split] 
test = dataset[split:]
plt.figure(figsize=(12,8))
plt.plot(train.Month, train.Sales, label='Train')
plt.plot(test.Month, test.Sales, label='Test')
plt.xticks(dataset["Month"], dataset["Month"], rotation='vertical')
plt.legend(loc='best')
plt.title("Train Test Split")
plt.show()
```
