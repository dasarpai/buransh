```python
import os
import google_auth_oauthlib.flow
import googleapiclient.discovery
import googleapiclient.errors


def get_authenticated_service():
    # Set up credentials
    scopes = ["https://www.googleapis.com/auth/youtube.readonly"]
    client_secrets_file = r"client_secret_913981478567-kv7vsikg270e76b9irdul1kesh56bbb3.apps.googleusercontent.com.json"  # Replace with your client secrets file

    flow = google_auth_oauthlib.flow.InstalledAppFlow.from_client_secrets_file(
        client_secrets_file, scopes
    )
    credentials = flow.run_local_server(port=0)
    youtube = googleapiclient.discovery.build("youtube", "v3", credentials=credentials)
    return youtube
```

```python
def get_subscribed_channels(youtube):
    subscribed_channels = []

    # Get first page of subscribed channels
    request = youtube.subscriptions().list(
        part="snippet",
        mine=True,
        maxResults=50
    )

    while request is not None:
        response = request.execute()
        for item in response.get("items", []):
            channel_title = item["snippet"]["title"]
            channel_id = item["snippet"]["resourceId"]["channelId"]
            subscribed_channels.append((channel_title, channel_id))

        request = youtube.subscriptions().list_next(request, response)

    return subscribed_channels
```

```python
# Get authenticated service
youtube = get_authenticated_service()
```

```python


# Get subscribed channels
subscribed_channels = get_subscribed_channels(youtube)

# Print the list of subscribed channels
for channel_title, channel_id in subscribed_channels:
    print(f"Title: {channel_title}")
    print(f"Channel ID: {channel_id}")
    print("--------------------------")
```
